```python
import os
import numpy as np
from PIL import Image
from mtcnn import MTCNN
```

```python
detector = MTCNN()
```

```python
def extractFace(image: Image):

    faces = detector.detect_faces(np.array(image))

    faces = sorted(faces, key=lambda x: x['confidence'], reverse=True)

    try:
        x, y, w, h = faces[0]['box']
    except:
        return image

    return image.crop((x, y, x + w, y + h))
    
```

```python
source_path = "../dataset/fsktm_dataset/"
target_path = "../dataset/fsktm_cropped_dataset/"

for root, _, files in os.walk(source_path):
    folder = root.split('/')[-1]
    target_folder = os.path.join(target_path, folder)

    if not os.path.isdir(target_folder):
        os.makedirs(target_folder)

    for file in files:
        path = os.path.join(root, file)

        img = Image.open(path)

        cropped = extractFace(img)

        cropped.save(os.path.join(target_folder, file))

        
```
